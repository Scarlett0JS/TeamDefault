```python
# 절대 절대 손대지 말것
from urllib import parse
import pyautogui
import string
import time
import json
import os

class MacroApp:
    def __init__(self):
        self.script_li = ["Python", "Java", "HTML", "JavaScript"]
        self.font_li = ['JetBrains Mono', 'D2coding', 'Noto Sans', 'Consolas', 'Fira code']
        self.script_X = 160 # 마우스로 첫번째 스크립트 클릭
        self.script_Y = 60
        self.font_XY = (580,611) # 마우스로 폰트 위치 클릭
        self.first_y = 88
        self.flag = True
        self.slackCnt = 0
        self.destList = []
    
    def checkFile(self, filename, p, imgPath):
        if filename + ".png" not in os.listdir(imgPath):
            return filename
        else:
            return self.checkFile(filename + "_" + str(p), p+1, imgPath)
    
    def fontChoose(self, imgPath, themename, themeinfo, themeurl, script):
        for font in self.font_li:
            pyautogui.hotkey('ctrl', ',')
            time.sleep(0.3)
            pyautogui.moveTo(*self.font_XY)
            pyautogui.doubleClick()
            pyautogui.click()
            time.sleep(0.1)
            pyautogui.press('delete')
            pyautogui.typewrite(font)
            time.sleep(0.5)
            pyautogui.hotkey('ctrl', 'w')
            time.sleep(3)
            
            imgFilename = self.checkFile(parse.quote(f"{themename}_{themeinfo}_{script}_{font}"), 1, imgPath)    
                
            self.destList.append({"name" : themename, "info" : themeinfo, "url" : themeurl, "imgpath" : f"{imgFilename}.png"})
            
            pyautogui.screenshot(os.path.join(imgPath, f"{imgFilename}.png"))
    
    def themeChoose(self):
        pyautogui.hotkey('ctrl', 'k', 't')
        time.sleep(0.5)
        if self.flag:
            # 항상 첫번째 테마가 선택된 채로 시작해야됨
            self.flag = False
        else:
            pyautogui.press('down')
            pyautogui.press('enter')
        time.sleep(2)    
    
    def startApp(self, imgPath, JsonData):
        pyautogui.hotkey('ctrl', 'b') # 탐색창 (활성화 상태로 시작 가정)
        for themename, themeinfo, themeurl in JsonData:
            # themeChoose
            self.themeChoose()
            
            # scripChoose
            for script in self.script_li:
                pyautogui.moveTo(self.script_X, self.script_Y)
                pyautogui.click()
                # fontChoose
                self.fontChoose(imgPath, themename, themeinfo, themeurl, script)
                self.script_X += 210 # jump cursor
            self.script_X = 160
            
        with open("reFinal_ThemeInfoImg.json", 'w', encoding="utf-8") as f2:
            json.dump(self.destList, f2, ensure_ascii=False)           
        
    def themeInfo(self, themeNameLi):
        for i in range(len(themeNameLi)):
            pyautogui.hotkey('ctrl', 'k', 't')
            
            if i < 12:
                pyautogui.moveTo(1305, self.first_y)
                self.first_y += 27
            else:
                pyautogui.moveTo(1305, self.first_y)
            
            time.sleep(0.5)
            pyautogui.click()
            time.sleep(0.2)
            
            pyautogui.moveTo(175, 105)
            pyautogui.doubleClick()
            pyautogui.click()
            time.sleep(0.1)
            pyautogui.hotkey('ctrl', 'c')
            
            pyautogui.moveTo(630,60)
            pyautogui.click()
            time.sleep(0.1)
            pyautogui.hotkey('ctrl', 'v')
            pyautogui.press('enter')
            
            if i < len(themeNameLi)-1:
                pyautogui.hotkey('ctrl', 'k', 't')
                pyautogui.press('down')
                pyautogui.press('enter')
                
    def makeThemeInfo(self):
        
        themeNameLi = self.makeThemeNameLi()
        with open("reThemeInfo.txt", 'r', encoding="utf-8") as f:
            themeInfoLi = [line.strip() for line in f.readlines()]
        
        urlLi = ["https://marketplace.visualstudio.com/items?itemName=" + i.split(":")[1] for i in themeInfoLi]
        print(len(themeNameLi), len(themeInfoLi), len(urlLi))
        
        if len(themeNameLi) == len(themeInfoLi) == len(urlLi):
            data = [{"name" : name, "info" : info.split(":")[1].split(".")[0], "url" : url} for name, info, url in zip(themeNameLi, themeInfoLi, urlLi)]
            with open("reFinal_ThemeInfo.json", 'w', encoding="utf-8") as f:
                json.dump(data, f, ensure_ascii=False)
        else:
            print("Not Match...")
        
    def makeThemeNameLi(self):
        with open("reThemeNamsText.txt", 'r', encoding="utf-8") as f:
            # themeNames = [line.strip().translate(str.maketrans('', '', string.punctuation)) for line in f.readlines()]
            themeNames = [line.strip().replace("/", "&") for line in f.readlines()]
            f.close()
        return themeNames
    
if __name__ == "__main__":
    test = MacroApp()
    
    # with open("reThemeNamsText.txt", 'r', encoding="utf-8") as f:
    #     themeNameLi = [i.strip() for i in f.readlines()]
    
    # test.themeInfo(themeNameLi)
    
    # test.makeThemeInfo() 

    with open("reFinal_ThemeInfo.json", 'r', encoding="utf-8") as f:
        data = json.load(f)
        f.close()

    JsonData = [(i['name'], i['info'], i['url']) for i in data]
    # print(len(JsonData))
    
    imgPath = os.path.join(os.getcwd(), 'ThemeImg')
    if not os.path.exists(imgPath):
        os.mkdir(imgPath)
        
    pyautogui.hotkey('ctrl', 's')
    test.startApp(imgPath, JsonData)
```
